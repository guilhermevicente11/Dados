```python
import pandas as pd
```

```python
df = pd.read_csv('/churn.csv')
df
```

```python
#contar a quanitdade de clientes que saíram
df['Exited'].value_counts()
```

```python
df[["EstimatedSalary"]]
```

```python
df.describe().round(2)
```

```python
df.corr().round(2)
```

```python
#media de idade dos clientes
df[["Age"]].mean()
```

```python
#media de idade do pessoal que saiu
saida_pessoal = df.loc[df.Exited == 1]
saida_idade = saida_pessoal[["Age","Exited"]]
saida_idade.mean()
```

```python
#media idade do pessoal que fica
fica_pessoal = df.loc[df.Exited == 0]
fica_idade = fica_pessoal[["Age","Exited"]]
fica_idade.mean()
```

```python
#media salario do pessoal que sai
saida_pessoal = df.loc[df.Exited == 1]
saida_salario = saida_pessoal[["EstimatedSalary","Exited"]]
saida_salario.mean()
```

```python
#media salario do pessoal que fica
fica_pessoal = df.loc[df.Exited == 0]
fica_salario = fica_pessoal[["EstimatedSalary","Exited"]]
fica_salario.mean()
```

```python
#score medio dos clientes
df[["CreditScore"]].mean()
```

```python
#score medio dos clientes que saem
fica_pessoal = df.loc[df.Exited == 0]
fica_score = fica_pessoal[["CreditScore","Exited"]]
fica_score.mean()
```

```python
#score medio dos clientes que ficam
saida_pessoal = df.loc[df.Exited == 1]
saida_score = saida_pessoal[["CreditScore","Exited"]]
saida_score.mean()
```

```python
#país que mais sai cliente
saida_pessoal = df.loc[df.Exited == 1]
saida_pais = saida_pessoal[["Exited","Geography"]]
saida_pais.hist()
```
